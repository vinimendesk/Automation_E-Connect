```python
import pandas as pd
import openpyxl
```

```python
file_path = r"C:/Users/vini7\Desktop/Projetos Automações Planilhas/Mariane/Projeto Automação Planilha Renovação/Automation_E-Connect/outputRenovacao.txt"
data = []

with open(file_path, "r", encoding="utf-8") as file:
    for line in file:
        parts = line.split(",")
        parceiro = parts[0].strip()
        index = int(parts[1].strip())
        data.append((index, parceiro))

df_parceiro = pd.DataFrame(data, columns=["index", "parceiros"]).set_index("index")
df_parceiro.head()
```
