```python
# Importando bibliotecas.
import pandas as pd
import pyautogui as pygui
import pyperclip
import time
import re
```

```python
# Abrindo arquivo xlsx.
arquivo = pd.read_excel(r"C:/Users/vini7/Desktop/Projetos Automações Planilhas/Mariane/Projeto Automação Planilha Renovação/Automation_E-Connect/Planilhas/RenovacaoAbr25Erro.xlsx")
arquivoPedidoExt = arquivo["pedidoExt"]
```

```python
arquivo
```

```python
# Regex para unidade.
unidadeString = re.compile(r'UNIDADE')
# Regex para analitica.
analiticaString = re.compile(r'SEDE')
# Regex para Parceiros
parceirosString = re.compile(r'Parceiros')
# Verificação PENDENTE.
pendenteString = "PENDENTE"
```

```python
arquivo["parceiro"] = ""
```

```python
# Função Ar Connect (Use na página 90%)
def autoArConnect(codigo):
    
    time.sleep(2)

    # Clica no botão filtro.
    pygui.click(x=151, y=187, button="left")
    time.sleep(1)

    # Clica no TextField.
    pygui.click(x=532, y=430, button="left")
    time.sleep(1)

    # Seleciona todo o texto.
    pygui.hotkey("ctrl", "a")
    time.sleep(0.5)  # Espera um pouco para garantir a seleção

    # Escreve o código.
    pygui.write(str(codigo))  # Garante que o código seja uma string
    time.sleep(0.5)  # Espera um pouco após escrever

    # Clica em pesquisar.
    pygui.click(x=877, y=545, button="left")
    time.sleep(4)  # Espera a página carregar

    # Seleciona editar informação.
    pygui.click(x=1260, y=242, button="left")
    time.sleep(5)

    # Seleciona o campo "Tipo Movimentação" ou "PENDENTE" para saber se deve ou não scrollar a página.
    pygui.click(x=110, y=330, button="left")
    # Seleciona o texto.
    pygui.hotkey("ctrl", "a")
    time.sleep(0.5)
    # Copia o texto.
    pygui.hotkey("ctrl", "c")
    time.sleep(0.5)
    #Clica fora da caixa.
    pygui.click(x=459, y=349, button="left")
    time.sleep(0.5)

    pendente = pyperclip.paste()
    
    # Verifica se há o status pendente.
    if pendente == pendenteString:

       
        pygui.press("down")
        pygui.press("down")
        pygui.press("down")
        pygui.press("down")
        pygui.press("down")
        pygui.press("down")

        time.sleep(0.5)

        # Seleciona o texto da "Unidade".
        pygui.click(x=155, y=700, button="left")
        time.sleep(0.5)
        # Seleciona o texto.
        pygui.hotkey("ctrl", "a")
        time.sleep(0.5)
        # Copia o texto.
        pygui.hotkey("ctrl", "c")
        time.sleep(0.5)    

        pygui.scroll(700)
        time.sleep(0.5)
        
    else: 

        pygui.press("down")
        pygui.press("down")
        time.sleep(0.5)        

        # Seleciona o texto da "Unidade".
        pygui.click(x=143, y=689, button="left")
        time.sleep(0.5)
        # Seleciona o texto.
        pygui.hotkey("ctrl", "a")
        time.sleep(0.5)
        # Copia o texto.
        pygui.hotkey("ctrl", "c")
        time.sleep(0.5)

    # Valor da unidade.
    unidade = pyperclip.paste()

    # Verifica se há "analitica" na string para procurar o nome do parceiro.
    if analiticaString.search(unidade):

        # Clica no campo "Parceiro indicador".
        pygui.click(x = 192, y = 513, button="left")
        time.sleep(0.5)
        # Seleciona o texto.
        pygui.hotkey("ctrl", "a")
        time.sleep(0.5)
        # Copia o texto.
        pygui.hotkey("ctrl", "c")
        time.sleep(0.5)
        # Valor da unidade.
        unidade = pyperclip.paste()

        pygui.scroll(700)
        time.sleep(0.5)
    else:
        pygui.scroll(700)
        time.sleep(0.5)


    # Clica em voltar.
    pygui.click(x=1316, y=150, button="left")
    time.sleep(3)
    #print(unidade)

    return str(unidade)

#autoArConnect("74158055")
```

```python
time.sleep(1)
# Substituindo/Adicionando parceiro.
for index, codigo in enumerate(arquivoPedidoExt):
    
    
    try:
        # Atribui o valor da unidade encontrado na automação.
        parceiro = autoArConnect(int(codigo))
        parceiro = re.sub(unidadeString, '', parceiro).strip()
        print(f"parceiro: {parceiro}, index: {index}")
        # Atribui o valor em maiúsculo ao indice da planilha.
        arquivo.at[index, "parceiro"] = parceiro.upper()
    except Exception as e:
        print(f"Erro ao salvar o arquivo {e}, no indice: {index}")
    

arquivo
arquivo.to_excel("RenovacaoAbr25erro.xlsx", index=False)
```
